Машинный эпсилон

### 🔬 Машинный эпсилон (ε): предел точности вычислений с плавающей запятой

**Что такое машинный эпсилон?**  
Минимальное число $\varepsilon > 0$, такое что $1.0 + \varepsilon > 1.0$ в арифметике с плавающей запятой.

**Почему это важно?**
- Определяет **предел относительной точности** вычислений (~15-16 знаков для float64)
- Объясняет, почему `0.1 + 0.2 != 0.3` в компьютере
- Задаёт нижнюю границу шага $h$ в численных методах (дифференцирование, интегрирование)

**Стандарт IEEE 754:**
| Тип | Размер | ε (машинный эпсилон) | Примерный диапазон |
|-----|--------|---------------------|-------------------|
| **float16** | 16 бит | ~$10^{-3}$ | $[6\cdot10^{-8}, 6\cdot10^{4}]$ |
| **float32** | 32 бит | ~$10^{-7}$ | $[10^{-38}, 10^{38}]$ |
| **float64** | 64 бит | ~$10^{-16}$ | $[10^{-308}, 10^{308}]$ |



````python
# === 0. Импорт библиотек ===
import sys
import numpy as np

# === 1. Эталонные значения ε из библиотек ===
print("📋 МАШИННЫЙ ЭПСИЛОН (ε) — минимальное число, для которого 1+ε > 1:\n")

# Стандартные значения для разных типов данных
print(f"{'Тип данных':<15} | {'Машинный ε':>15} | {'Значимых цифр':>15}")
print("-" * 52)

# Python (всегда float64)
eps_py = sys.float_info.epsilon

print(f"{'sys.float_info':<15} | {eps_py:15.2e} | {-np.log10(eps_py):>14.0f} знаков")

# NumPy: основные типы
for dtype, name in [(np.float64, 'float64'), (np.float32, 'float32'), (np.float16, 'float16')]:
    eps_np = np.finfo(dtype).eps
    digits = -np.log10(eps_np)
    # ИСПРАВЛЕНО: округление до целых через .0f
    print(f"{name:<15} | {eps_np:15.2e} | {digits:>14.0f} знаков")

# === 2. Вычисление ε "в лоб" — алгоритмическая демонстрация ===
print(f"\n🔍 Вычисляем ε для float64 алгоритмически:")
print("   Алгоритм: делим ε пополам, пока 1+ε не перестанет отличаться от 1")

eps_computed = 1.0
iteration = 0

# Цикл деления пополам: ищем границу, где арифметика "не видит" разницы
while (1.0 + eps_computed/2) > 1.0:
    eps_computed /= 2
    iteration += 1

print(f"   Итераций: {iteration}")
print(f"   🏎️  Вычисленный ε = {eps_computed:.2e}")

# === 3. Проверка: демонстрируем "границу видимости" ===
print(f"\n✅ Проверка результата:")
print(f"   1.0 + ε/2 = {1.0 + eps_computed/2:.17f}  → равно 1.0? {1.0 + eps_computed/2 == 1.0}")
print(f"   1.0 + ε   = {1.0 + eps_computed:.17f}  → равно 1.0? {1.0 + eps_computed == 1.0}")
print(f"   1.0 + 2ε  = {1.0 + 2*eps_computed:.17f}  → видно отличие ✓")

# === 4. Практическая иллюстрация: почему это важно ===
print(f"\n💡 Практический пример:")
a = 1.0
b = 1.0 + eps_computed/2  # Меньше ε — отличие теряется!
print(f"   a = {a:.17f}")
print(f"   b = 1.0 + ε/2 = {b:.17f}")
print(f"   a == b ? {a == b}  ← ⚠️ Компьютер 'не видит' разницы!")
````

````python
# === 🎯 Демонстрация: потеря малого слагаемого ===
import numpy as np

# Числа разных порядков
big = 1e16      # Большое число
small = 1.0     # Малое число (на 16 порядков меньше!)



# Сценарий 1: big + small
s1 = big + small
print(f"big + small         = {s1:.1f}")
print(f"  (должно быть 10000000000000001.0)")

# Сценарий 2: big + small + small
s2 = big + small + small
print(f"big + small + small = {s2:.1f}")
print(f"  (должно быть 10000000000000002.0)")

# Проверяем разность: если small "потерялся", разность будет 0
print(f"\nРазность s2 - s1    = {s2 - s1:.1f}")
print(f"  (должно быть 1.0, но получили {s2 - s1})")

# === 🔬 Демонстрация "исчезновения" в явном виде ===
print(f"\n=== 🎯 Сдвигаемся в масштаб: (big + small) - big ===")
big2 = 1e16
small2 = 1.0
result = (big2 + small2) - big2

print(f"small2              = {small2:.1f}")
print(f"(big2 + small2) - big2 = {result:.1f}")
print(f"  ⚠️ Малое число 'исчезло' при сложении с большим!")
````

### Очень простой демонстрационный пример с «плохим» и «хорошим» порядком суммирования:

### 🔄 Порядок суммирования: почему он важен для точности?

**Проблема:** При последовательном сложении `s += x` малые слагаемые могут **теряться**, если текущая сумма `s` уже велика.

**Почему?**  
Если $|x| < |s| \cdot \varepsilon$, то $\text{fl}(s + x) = s$ — малое число не влияет на результат!

**Решение:** Складывать числа **от меньшего к большему** — так малые вклады успевают "накопиться" до значимой величины.

> 💡 **Эксперимент:** 10 чисел по $10^{16}$ + 1 000 000 чисел по $1.0$.  
> Теоретическая сумма: $10^{17} + 10^6 = 10000000000000000

````python
# === 🎯 Демонстрация: влияние порядка суммирования на точность ===
import numpy as np

# Исходные данные: 10 больших чисел + 1 миллион малых
values = [1e16] * 10 + [1.0] * 1_000_000

# Теоретическая точная сумма (в математике)
exact_sum = 10 * 1e16 + 1_000_000 * 1.0


# --- ❌ ПЛОХОЙ порядок: как есть (большие числа первыми) ---
# Малые слагаемые добавляются к уже огромной сумме → теряются из-за ε
s_bad = 0.0
for v in values:
    s_bad += v

# --- ✅ ХОРОШИЙ порядок: сначала малые, потом большие ---
# Малые числа складываются между собой → накапливаются → становятся значимыми
s_good = sum(sorted(values))  # sorted() по возрастанию: [1.0, 1.0, ..., 1e16]

# --- Сравнение результатов ---
print("=== 📊 Результаты суммирования ===")
print(f"Плохая сумма (исходный порядок): {s_bad:.1f}")
print(f"Хорошая сумма (по возрастанию) : {s_good:.1f}")
print(f"Теоретическая сумма            : {exact_sum:.1f}")

print(f"\n=== 🔍 Погрешности ===")
err_bad = abs(s_bad - exact_sum)
err_good = abs(s_good - exact_sum)
print(f"Ошибка плохого порядка : {err_bad:.2e}  ({err_bad/exact_sum*100:.2e}%)")
print(f"Ошибка хорошего порядка: {err_good:.2e}  ({err_good/exact_sum*100:.2e}%)")


# === 🔬 Бонус: почему сортировка помогает? ===
print(f"\n=== 💡 Механика потерь ===")
eps = np.finfo(float).eps
big = 1e16
print(f"Машинный ε              = {eps:.2e}")
print(f"Порог потери для {big:.1e}: |x| < {big*eps:.1f}")
print(f"Наше малое слагаемое    = 1.0")
print(f"1.0 < {big*eps:.1f} ? {1.0 < big*eps} → ⚠️ теряется при добавлении к {big:.1e}")
````

Проблема обычного суммирования: при сложении большого количества чисел ошибки округления накапливаются, особенно если числа сильно различаются по порядку.
​

Идея Кахана: ведётся учёт «потерянной» при округлении части суммы и добавляется она в следующий шаг. Это компенсирует ошибки округления


s ← 0     # основная сумма

c ← 0     # компенсация (потерянная часть)




````python
# === 🛡️ Алгоритм Кахана: компенсированное суммирование ===
import numpy as np

def kahan_sum(arr: np.ndarray) -> float:
    """
    Алгоритм Кахана для точного суммирования чисел с плавающей запятой.

    Принцип работы:
    ---------------
    На каждом шаге мы отслеживаем ошибку округления и компенсируем её
    при добавлении следующего слагаемого.

    Переменные:
    - s: текущая сумма (sum)
    - c: накопленная ошибка округления (compensation)
    - y: слагаемое с учётом компенсации
    - t: новая сумма после добавления
    """
    # Инициализируем в float64 для явной точности
    s = np.float64(0.0)  # Сумма
    c = np.float64(0.0)  # Компенсация (накопленная ошибка)

    for x in arr:
        # Шаг 1: Учитываем накопленную ошибку из предыдущего шага
        y = np.float64(x) - c  # "Очищенное" слагаемое

        # Шаг 2: Добавляем к сумме (здесь происходит округление!)
        t = s + y  # Новая сумма

        # Шаг 3: Вычисляем, сколько потеряли при округлении
        # (t - s) — это то, что фактически добавилось к s
        # y — это то, что мы хотели добавить
        # Разница — это потерянная часть
        c = (t - s) - y  # Новая ошибка для компенсации

        # Шаг 4: Обновляем сумму
        s = t

    return s


# === 🧪 Эксперимент: сравнение с обычным суммированием ===
# Создаём тестовые данные: 1 большое + 1 миллион малых
values = [1e16] * 1 + [0.5] * 1_000_000 +   [0.5] * 1_000_000

# Теоретически точная сумма (математика)
exact_sum = 1e16 + 1_000_000 * 1.0

# Сравнение трёх методов
print("=== 📊 Сравнение методов суммирования ===")
print(f"{'Метод':<25} | {'Результат':>22} | {'Ошибка':>12}")
print("-" * 65)


# 1. NumPy sum (использует частичную компенсацию)
sum_numpy = np.sum(values)
err_numpy = abs(sum_numpy - exact_sum)
print(f"{'np.sum()':<25} | {sum_numpy:22.2f} | {err_numpy:12.6f}")

# 2. Алгоритм Кахана (наша реализация)
sum_kahan = kahan_sum(np.array(values))
err_kahan = abs(sum_kahan - exact_sum)
print(f"{'Kahan sum ':<25} | {sum_kahan:22.2f} | {err_kahan:12.6f}")

````

Надо сказать. что в numpy  для float типа используется алгоритм частичного парного суммирования.

Если же нужно еще более стабилное вычисление суммы, то можно использовать math.fsum.  Python-функция math.fsum() представляет собой высокоточную методику суммирования чисел с плавающей точкой. Она использует специальный алгоритм ( Shewchuk в Python), который минимизирует потерю точности, характерную для стандартной арифметики с плавающей точкой.

## ⚡ Бенчмарк методов суммирования: скорость vs. точность на 100 млн чисел

**Цель:** Сравнить три подхода к суммированию на реалистичном объёме данных:

| Метод | Алгоритм | Точность | Скорость | Когда использовать |
|-------|----------|----------|----------|-------------------|
| **`sum()`** | Последовательное сложение | ⚠️ Средняя | 🟡 Средняя | Прототипы, небольшие данные |
| **`np.sum()`** | Векторизация + SIMD + многопоточность | ⚠️ Зависит от порядка | 🚀 Быстро | Большие массивы, однородные данные |
| **`math.fsum()`** | Алгоритм Шьючука (точное округление) | 🏆 Максимальная | 🐌 Медленно | Финансы, научные расчёты, где важна каждая цифра |

**Тестовый сценарий:**  
$3 \times 1.0 + 100\,000\,000 \times 10^{-8} = 3 + 1 = 4$  
Малые слагаемые в $10^8$ раз меньше больших — идеальная проверка на потерю значимости!


````python
# === ⚡ Бенчмарк: сравнение методов суммирования ===
import numpy as np
import math
import time

def benchmark_sums(n_big=3, n_small=100_000_000):
    """
    Сравнивает точность и скорость трёх методов суммирования.

    Параметры:
    ----------
    n_big : int
        Количество больших чисел (1.0)
    n_small : int
        Количество малых чисел (1e-8)
    """
    print(f"\n{'='*70}")
    print(f"🚨 BENCHMARK: {n_big}×1.0 + {n_small:,}×1e-8")
    print(f"{'='*70}")

    # === 1. Подготовка данных ===
    # Большие числа идут первыми — худший порядок для точности!
    numbers_desc = np.full(n_big, 1.0, dtype=np.float64)
    numbers_asc = np.full(n_small, 1e-8, dtype=np.float64)

    # Конкатенируем: сначала большие, потом малые (провоцируем потерю точности)
    numbers = np.concatenate([numbers_desc, numbers_asc])

    # Теоретически точная сумма (математика, без округлений)
    s_theory = n_big * 1.0 + n_small * 1e-8
    print(f"🎯 Теоретическая сумма: {s_theory:.10f}\n")

    # === 2. Словарь методов для тестирования ===
    methods = {
        'sum()': lambda x: sum(x),           # Python built-in
        'np.sum()': lambda x: np.sum(x),     # NumPy векторизованный
        'math.fsum()': lambda x: math.fsum(x) # Точное суммирование
    }

    results = {}
    times = {}

    # === 3. Замер времени и точности для каждого метода ===
    print(f"{'Метод':<12} | {'Результат':>18} | {'Ошибка Δ':>12} | {'Время':>8} | {'Статус'}")
    print("-" * 70)

    for name, func in methods.items():
        start = time.perf_counter()
        result = func(numbers)
        end = time.perf_counter()

        results[name] = result
        times[name] = (end - start) * 1000  # Время в миллисекундах
        delta = abs(result - s_theory)      # Абсолютная ошибка

        # Статус: ✅ если ошибка близка к машинному эпсилону
        status = "✅" if delta < 1e-10 else "❌"
        print(f"{name:<12} | {result:18.15f} | {delta:12.2e} | {times[name]:6.2f}мс | {status}")

    # === 4. Итоговая сводка ===
    print(f"\n{'='*70}")
    print(f"📊 ИТОГИ:")

    # Находим лучший по точности и по скорости
    best_accuracy = min(results.keys(), key=lambda k: abs(results[k] - s_theory))
    best_speed = min(times.keys(), key=lambda k: times[k])

    print(f"• 🎯 Наилучшая точность: {best_accuracy}")
    print(f"• 🚀 Наивысшая скорость:  {best_speed}")
    print(f"{'='*70}\n")

    return results, times

# === 🚀 Запуск бенчмарка ===
# Внимание: n_small=100_000_000 может занять ~10-30 секунд на math.fsum()
benchmark_sums(n_big=3, n_small=100_000_000)
````


### 📈 Анализ результатов:

| Метод | Ошибка | Время | Почему так? |
|-------|--------|-------|-------------|
| **`sum()`** | ~$6\cdot10^{-8}$ | ~30 мс | Детерминированный порядок, но нет компенсации ошибок |
| **`np.sum()`** | ~$5\cdot10^{-13}$ | ~8 мс | ⚡ Векторизация + SIMD, но порядок операций недетерминирован → ошибка тоже присутствует |
| **`math.fsum()`** | **0** | ~850 мс | 🏆 Алгоритм Шьючука: точное округление на каждом шаге, но в 25-100× медленнее |

## Разбор структуры числа в формате IEEE 754 double precision

````python
import struct
import numpy as np

pi = np.pi
# Биты double
bits = struct.unpack('Q', struct.pack('d', pi))[0]

print(f"π в double (bin): {bits:064b}")

# Разбор по полям
sign = (bits >> 63) & 1
exponent = (bits >> 52) & 0x7FF
mantissa = bits & 0xFFFFFFFFFFFFF

print(f"\nЗНАК:     {sign} ({'+' if sign==0 else '-'})")
print(f"ПОРЯДОК:  {exponent} (истинный = {exponent-1023})")
print(f"МАНТИССА: {mantissa:064b} (скрытая 1.)")
print(f"МАНТИССА: {mantissa} (скрытая 1.)")
````



Получаем:

ЗНАК = 0  
$(-1)^{0}$ = +1 ✓  

ПОРЯДОК = 1024  
$1024 - 1023 = 1 \rightarrow 2^{1}$ ✓  

МАНТИССА  

Перевод мантиссы в десятичные дроби (первые несколько бит):  

$$
1.0000000000001001001000011111101101010100010001000010110100011000_2 \\

$$

$$

$$

````python
# Собираем обратно
sign, exp, mant = 0, 1024, 2570638124657944
true_exp = exp - 1023  # = 1

# Мантисса как float: 1 + mantissa/(2^52)
mantissa_float = 1.0 + mant / (2**52)
pi_reconstructed = ((-1)**sign) * (2**true_exp) * mantissa_float

print(f"Восстановленное π: {pi_reconstructed}")
print(f"np.pi:              {np.pi}")
print(f"Равны:              {abs(pi_reconstructed - np.pi) < 1e-15}")
````

##  Умножение чисел

## 🧮 Устойчивое умножение: как избежать переполнения и исчезновения порядка?

**Проблема:** При перемножении многих чисел промежуточный результат может:
- **Переполниться** (→ `inf`), если множим много чисел > 1
- **Исчезнуть** (→ `0`), если множим много чисел < 1



````python
# === 🧮 Устойчивые алгоритмы умножения ===
import numpy as np
import math

def regular_product(nums):
    """Наивное умножение — быстро, но может переполниться"""
    res = 1.0
    for n in nums:
        res *= n
    return res

def log_product(nums, return_log=False):
    """
    Логарифмическое умножение.

    Параметры:
    ----------
    return_log : bool
        Если True, возвращаем log(произведения) — это всегда безопасно!
    """
    nums = np.array(nums, dtype=np.float64)
    if np.any(nums == 0):
        return 0.0 if not return_log else -np.inf

    sign = np.prod(np.sign(nums))
    log_sum = np.sum(np.log(np.abs(nums)))

    # Если просят логарифм — возвращаем сразу (без exp!)
    if return_log:
        return log_sum if sign > 0 else np.nan

    # Проверка: exp() переполнится при log_sum > 709
    if log_sum > 709:
        return np.inf
    elif log_sum < -745:
        return 0.0

    return sign * np.exp(log_sum)


# === 🧪 Пример 1: Переполнение (regular fails, log в логах работает) ===
print("="*60)
print("🧪 Пример 1: Переполнение (10^310 > 10^308)")
print("="*60)

test1 = [10.0] * 310  # 10^310 — больше предела float64 (~10^308)

reg = regular_product(test1)
log_val = log_product(test1)
log_sum = log_product(test1, return_log=True)  # ✅ Безопасно!

print(f"regular_product: {reg}  ❌ overflow (inf)")
print(f"log_product:     {log_val}  ⚠️ тоже inf (exp(713) > 709)")
print(f"\n✅ РЕШЕНИЕ: остаться в лог-пространстве!")
print(f"   log(произведения) = {log_sum:.2f}")
print(f"   Это корректно! (10^310 → log = 310·log(10) ≈ 713)")


# === 🧪 Пример 2: Исчезновение (regular fails, log works) ===
print(f"\n{'='*60}")
print("🧪 Пример 2: Исчезновение (0.01^200 < 10^-308)")
print("="*60)

test2 = [0.01] * 200  # 0.01^200 = 10^-400 — меньше предела float64

reg = regular_product(test2)
log_val = log_product(test2)

print(f"regular_product: {reg}  ❌ underflow (0.0)")
print(f"log_product:     {log_val:.4e}  ✅ корректно!")
print(f"Теоретически:    0.01^200 = 10^-400")


# === 🧪 Пример 3: Накопление ошибки (log точнее) ===
print(f"\n{'='*60}")
print("🧪 Пример 3: Накопление ошибки (10 000 умножений)")
print("="*60)

test3 = [1.001] * 10000  # 1.001^10000 ≈ 22026

reg = regular_product(test3)
log_val = log_product(test3)
theory = 1.001**10000

err_reg = abs(reg - theory) / theory
err_log = abs(log_val - theory) / theory

print(f"regular_product: {reg:.4f}  (ошибка: {err_reg:.2e})")
print(f"log_product:     {log_val:.4f}  (ошибка: {err_log:.2e})")
print(f"✅ log_product точнее в {err_reg/err_log:.0f} раз!")


# === 🧪 Пример 4: ML-сценарий (произведение вероятностей) ===
print(f"\n{'='*60}")
print("🧪 Пример 4: ML — произведение 1000 вероятностей")
print("="*60)

probs = [0.5] * 1000  # 0.5^1000 ≈ 10^-301

reg = regular_product(probs)
log_val = log_product(probs)
log_likelihood = np.sum(np.log(probs))  # ✅ В ML используют это!

print(f"regular_product:   {reg:.4e}  ⚠️ риск underflow")
print(f"log_product:       {log_val:.4e}  ✅ надёжно")
print(f"\n🎯 В ML работают в логах:")
print(f"   log-likelihood = {log_likelihood:.2f}")
print(f"   (никогда не вычисляют произведение явно!)")

````

## 🔬 Эталонное вычисление с произвольной точностью (mpmath)

**Зачем это нужно?**  
Когда мы тестируем устойчивые алгоритмы, нам нужно знать **точное значение**, с которым сравнивать результаты float64.

**mpmath:** Библиотека для арифметики произвольной точности (software floating-point).

| Тип | Точность | Скорость | Когда использовать |
|-----|----------|----------|-------------------|
| **float64** | ~16 знаков | 🚀 Быстро | Основные вычисления |
| **mpmath (50 dps)** | 50 знаков | 🐌 Медленно (×100-1000) | ✅ Эталон для тестов, отладка |
| **mpmath (100+ dps)** | 100+ знаков | 🐢 Очень медленно | Критические научные расчёты |

> 💡 **Важно:** mpmath — это программная эмуляция, а не аппаратная арифметика. Используйте только для валидации!

````python
# === 🔬 Эталонное вычисление через mpmath (произвольная точность) ===
from mpmath import mp

# Настраиваем точность: 50 десятичных знаков (digits of precision)
mp.dps = 50

print(f"🎯 Вычисляем произведение с точностью {mp.dps} знаков:\n")

# Исходные данные (пример: 1000 чисел 1.001)
nums = [1.001] * 1000

# --- 1. Вычисление в mpmath ---
# Важно: конвертируем через str(), чтобы избежать ошибок float64 при передаче!
prod_mp = mp.mpf(1)
for num in nums:
    prod_mp *= mp.mpf(str(num))  # str(num) → точная десятичная запись

# --- 2. Сравнение с float64 методами ---
reg = regular_product(nums)
log_val = log_product(nums)

# --- 3. Вывод результатов ---
print("=== 📊 Сравнение с эталоном (mpmath) ===")
print(f"mpmath (50 знаков):  {prod_mp}")  # mpmath сам форматирует при print
print(f"\nfloat64 результаты:")
print(f"  regular_product:   {reg:.15e}  (ошибка: {abs(reg - float(prod_mp)):.2e})")
print(f"  log_product:       {log_val:.15e}  (ошибка: {abs(log_val - float(prod_mp)):.2e})")

# --- 4. Бонус: показываем, где float64 "отрезает" знаки ---
print(f"\n=== 🔍 Детали: сколько знаков совпадает? ===")

# ИСПРАВЛЕНО: используем mp.nstr() для форматирования mpmath чисел
mp_str = mp.nstr(prod_mp, 20)  # 20 значащих цифр
float_str = f"{float(prod_mp):.20g}"

print(f"mpmath : {mp_str}")
print(f"float64: {float_str}")

# Сравниваем первые 16 символов (предел float64)
mp_clean = str(mp_str).replace('.', '')[:16]
float_clean = float_str.replace('.', '')[:16]
print(f"         {'✓' if mp_clean == float_clean else '✗'} Совпадают первые ~16 знаков (предел float64)")
````

## Суммирование большого числа чисел

````python
# === 🔄 Тест: влияние порядка суммирования на точность ===
import numpy as np

def test_sum_order(dtype=np.float32):
    """
    Сравнивает точность суммирования ряда 1/k² в разном порядке.

    Параметры:
    ----------
    dtype : np.dtype
        Тип данных для вычислений (float32 или float64)
    """
    print(f"\n{'═'*70}")
    print(f"🔥 ТЕСТ: ПОРЯДОК СУММИРОВАНИЯ ({dtype.__name__.upper()})")
    print(f"{'═'*70}")

    # Тестируем на разных масштабах: n = 10⁴, 10⁵, 10⁶
    for n_exp in [4, 5, 6]:
        n = 10**n_exp

        # ✅ ХОРОШИЙ порядок: малые слагаемые первыми
        # Генерируем: 1/n², 1/(n-1)², ..., 1/1² (возрастание)
        x_good = 1.0 / ((np.arange(n)[::-1] + 1)**2).astype(dtype)

        # ❌ ПЛОХОЙ порядок: большие слагаемые первыми
        # Генерируем: 1/1², 1/2², ..., 1/n² (убывание)
        x_bad = 1.0 / ((np.arange(n) + 1)**2).astype(dtype)

        # "Эталон": считаем в float64 для сравнения
        s_exact = np.sum(x_good.astype(np.float64))

        # Наивное суммирование Python (слева направо)
        s_naive_good = sum(x_good)  # малые → большие
        s_naive_bad = sum(x_bad)    # большие → малые

        # Вывод результатов
        print(f"\n📊 n = {n:,} | Точное S ≈ {s_exact:.6f}")
        print(f"✅ Малые→Большие: {s_naive_good:.8f}  | Ошибка Δ = {abs(s_naive_good - s_exact):.2e}")
        print(f"❌ Большие→Малые: {s_naive_bad:.8f}  | Ошибка Δ = {abs(s_naive_bad - s_exact):.2e}")



# Запуск теста для float32 (наиболее наглядный случай)
test_sum_order(np.float32)
````

````python
import math

def fast_math_round(x, decimals=0):
    """Быстрое математическое округление"""
    multiplier = 10**decimals
    return math.copysign(1, x) * math.floor(abs(x) * multiplier + 0.5) / multiplier

print(fast_math_round(2.3456, 2))  # 2.35
print(fast_math_round(2.5, 0))     # 3.0
print(fast_math_round(-2.325, 2))  # -2.33
````
